```python
# Apply PCA
pca = PCA(n_components=len(X2.columns))  # Get all components
pca.fit(X2)

# Explained variance ratios
explained_variance = pca.explained_variance_ratio_

# Determine the number of components needed to explain sufficient variance (e.g., 80%)
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
n_components_needed = np.argmax(cumulative_variance >= 0.80) + 1  # First index where variance >= 80%

# Re-run PCA with selected number of components
pca_refined = PCA(n_components=n_components_needed)
X2_trans = pca_refined.fit_transform(X2)

# Component loadings (weights)
component_weights = np.abs(pca_refined.components_).sum(axis=0)

# Normalize weights to sum to 1
normalized_weights = component_weights / component_weights.sum()

#print(X2)
#print(normalized_weights)
#print(y2)

signs = np.sign(X2.corrwith(y2.squeeze(),axis=0)).fillna(1)  # +1 or -1 per variable

#print(signs)

X2_flipped = X2 * signs  # Element-wise multiplication (sign flip)

# Output the weights assigned to each variable
weights_df = pd.DataFrame({
    "Variable": X2.columns[:],
    "PCA_Weight": normalized_weights,
    "CorrelationSign": signs.values[:]
})

# Construct the new Geopolitical Distance Index (GDI)
X2_flipped["GDI"] = np.dot(X2_flipped, normalized_weights)

n_components_needed, weights_df
```

```python
def bai_perron_test(df, country_col='StateName2', year_col='year',
                                        value_col='adjusted_tradeflow',
                                        max_breaks=10, model='l2'):
    """
    Applies ruptures change point detection to each country's trade flow over time.
    
    Parameters:
        df (pd.DataFrame): Combined dataset
        country_col (str): Column for country name
        year_col (str): Column for year
        value_col (str): Column for the signal (e.g., adjusted_tradeflow)
        max_breaks (int): Max number of breakpoints to detect
        model (str): Model type for change point detection ("l2", "rbf", etc.)

    Returns:
        dict: Mapping from country to list of break years
    """
    break_dict = {}
    
    countries = df[country_col].unique()
    for country in countries:
        df_country = df[df[country_col] == country].sort_values(year_col)
        
        # Ensure data has enough years to detect breaks
        if df_country[year_col].nunique() < max_breaks + 2:
            continue

        # Get signal
        signal = df_country[value_col].values

        try:
            algo = rpt.KernelCPD(kernel="rbf").fit(signal.reshape(-1, 1))
            breaks = algo.predict(n_bkps=max_breaks)
            
            break_years = df_country[year_col].iloc[[b - 1 for b in breaks if b < len(df_country)]].tolist()
            break_dict[country] = break_years
        except Exception as e:
            print(f"Skipping {country} due to error: {e}")

    return break_dict
bai_perron_test(df_combined)
```

```python
def chow_test(df, dep_var='adjusted_tradeflow', indep_vars=['linear_gdi'], break_year=2008, log=True):
    df = df.copy()
    if log:
        df[dep_var]=np.log(df[dep_var])
    df = df.dropna(subset=[dep_var] + indep_vars + ['year'])  # Clean NAs
    df['year'] = pd.to_numeric(df['year'])

    # Split data before and after break
    df_pre = df[df['year'] < break_year]
    df_post = df[df['year'] >= break_year]

    if df_pre.empty or df_post.empty:
        raise ValueError("Insufficient data on one side of the break year.")

    # X and y for each segment
    X_pre = sm.add_constant(df_pre[indep_vars])
    y_pre = df_pre[dep_var]

    X_post = sm.add_constant(df_post[indep_vars])
    y_post = df_post[dep_var]

    X_full = sm.add_constant(df[indep_vars])
    y_full = df[dep_var]

    # Fit models
    model_pre = sm.OLS(y_pre, X_pre).fit()
    model_post = sm.OLS(y_post, X_post).fit()
    model_pooled = sm.OLS(y_full, X_full).fit()

    # RSS values
    RSS_pooled = np.sum(model_pooled.resid ** 2)
    RSS_pre = np.sum(model_pre.resid ** 2)
    RSS_post = np.sum(model_post.resid ** 2)

    k = X_full.shape[1]  # Number of parameters (incl. intercept)
    n1 = len(y_pre)
    n2 = len(y_post)

    # Chow F-statistic
    numerator = (RSS_pooled - (RSS_pre + RSS_post)) / k
    denominator = (RSS_pre + RSS_post) / (n1 + n2 - 2 * k)
    F_stat = numerator / denominator

    from scipy.stats import f
    p_value = 1 - f.cdf(F_stat, dfn=k, dfd=n1 + n2 - 2 * k)

    return {
        'break_year': break_year,
        'F_stat': F_stat,
        'p_value': p_value,
        'reject_H0': p_value < 0.05
    }

    
break_years = [_ for _ in range(1965,2024)]

results=[]
for country in df_combined['StateName2'].unique():
    if country!='Japan':continue
    df_country = df_combined[df_combined['StateName2'] == country]
    for year in break_years:
        try:
            result = chow_test(df_country, dep_var='adjusted_tradeflow', indep_vars=['linear_gdi'], break_year=year, log=True)
            result['country'] = country
            result['year'] = year
            results.append(result)
        except:
            continue

df_results = pd.DataFrame(results)
```

```python
import ruptures as rpt

def detect_breaks(series, model="l2", n_bkps=5):
    """
    Detect structural breaks in a 1D time series using ruptures.
    """
    algo = rpt.Pelt(model=model).fit(series)
    breaks = algo.predict(pen=3)  # or use n_bkps=n_bkps if fixed number
    return breaks

# Store break years per country
breaks_gdi = {}
breaks_trade = {}

for country in df_combined['StateName2'].unique():
    df_country = df_combined[df_combined['StateName2'] == country].sort_values('year')

    gdi_series = df_country["linear_gdi"].values
    trade_series = df_country["adjusted_tradeflow"].values
    years = df_country["year"].values

    if len(gdi_series) > 10:  # Only test if sufficient data
        gdi_breaks = detect_breaks(gdi_series)
        trade_breaks = detect_breaks(trade_series)

        # Convert to actual years
        breaks_gdi[country] = years[np.array(gdi_breaks[:-1])].tolist()
        breaks_trade[country] = years[np.array(trade_breaks[:-1])].tolist()

from collections import defaultdict

break_alignment_scores = defaultdict(int)

for country in breaks_gdi:
    gdi_years = set(breaks_gdi.get(country, []))
    trade_years = set(breaks_trade.get(country, []))

    aligned = [gy for gy in gdi_years if any(abs(gy - ty) <= 1 for ty in trade_years)]
    break_alignment_scores[country] = len(aligned)

import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
import matplotlib.pyplot as plt

# Build binary break matrix
all_years = sorted(df_combined["year"].unique())
country_break_matrix = pd.DataFrame(0, index=breaks_gdi.keys(), columns=all_years)

for country, years in breaks_gdi.items():
    for y in years:
        if y in country_break_matrix.columns:
            country_break_matrix.loc[country, y] = 1

# Cluster using hierarchical clustering
clustering = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='ward')
country_break_matrix["cluster"] = clustering.fit_predict(country_break_matrix.values)

# Visualize
sns.clustermap(country_break_matrix.drop(columns="cluster"), cmap="Blues", figsize=(36, 32))
plt.title("Country Clustering Based on GDI Break Patterns")
plt.show()
```

```python
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Features and target
features = ["linear_gdi", "dist", "gdp_o", "gdp_d", "pop_o", "pop_d"]
X = df_combined[features].dropna().values
y = np.log(df_combined["adjusted_tradeflow"].loc[df_combined[features].dropna().index].values)

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)

# Convert to tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)

# Neural net model
class TradeNN(nn.Module):
    def __init__(self, input_dim):
        super(TradeNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 32)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(32, 16)
        self.relu2 = nn.ReLU()
        self.output = nn.Linear(16, 1)

    def forward(self, x):
        x = self.relu1(self.fc1(x))
        x = self.relu2(self.fc2(x))
        return self.output(x)

model = TradeNN(input_dim=X_train.shape[1])
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Train
epochs = 500
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    output = model(X_train_tensor)
    loss = criterion(output, y_train_tensor)
    loss.backward()
    optimizer.step()

# Evaluate
model.eval()
y_pred = model(X_test_tensor).detach().numpy()
mse = ((y_pred - y_test_tensor.numpy())**2).mean()
print("MSE:", mse)
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import DataLoader, TensorDataset

# --- Step 1: Prepare time-series data (example for Singapore -> All Countries) ---
df_lstm = df_combined[df_combined["StateName1"] == "Singapore"].sort_values("year")
features = ["linear_gdi", "gdp_d", "pop_d", "adjusted_tradeflow"]
df_lstm = df_lstm[["year"] + features].dropna()

# --- Step 2: Normalize ---
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df_lstm[features])

# --- Step 3: Create sequences ---
def create_sequences(data, seq_len=5):
    xs, ys = [], []
    for i in range(len(data) - seq_len):
        x = data[i:i+seq_len]
        y = data[i+seq_len][-1]  # target: adjusted_tradeflow
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_len = 5
X_seq, y_seq = create_sequences(scaled_data, seq_len)
X_tensor = torch.tensor(X_seq, dtype=torch.float32)
y_tensor = torch.tensor(y_seq, dtype=torch.float32).view(-1, 1)

# --- Step 4: Train/Test Split ---
split = int(0.8 * len(X_tensor))
train_dataset = TensorDataset(X_tensor[:split], y_tensor[:split])
test_dataset = TensorDataset(X_tensor[split:], y_tensor[split:])
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)

# --- Step 5: LSTM Model ---
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size=50):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        _, (hn, _) = self.lstm(x)
        return self.fc(hn[-1])

model = LSTMModel(input_size=X_tensor.shape[2])
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# --- Step 6: Training ---
for epoch in range(100):
    model.train()
    for x_batch, y_batch in train_loader:
        pred = model(x_batch)
        loss = loss_fn(pred, y_batch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    if epoch % 10 == 0:
        print(f"Epoch {epoch} | Loss: {loss.item():.4f}")

# --- Step 7: Predict and Plot ---
model.eval()
preds = model(X_tensor).detach().numpy()
true_vals = y_tensor.numpy()

import matplotlib.pyplot as plt
plt.plot(range(len(true_vals)), true_vals, label="True")
plt.plot(range(len(preds)), preds, label="Predicted")
plt.title("LSTM Forecast: Adjusted Trade Flow")
plt.legend()
plt.show()
```
