```python
df_countries = df_names[['country_id','country']]
df_grav = df_gravity[["year","country_id_o","country_id_d","contig","distcap","dist","diplo_disagreement","comlang_off","comlang_ethno","comrelig","pop_o", "pop_d", "gdp_o", "gdp_d", "gdpcap_o", "gdpcap_d", "gdp_ppp_o", "gdp_ppp_d", "gdpcap_ppp_o", "gdpcap_ppp_d","fta_wto","tradeflow_comtrade_o","tradeflow_comtrade_d","tradeflow_baci","tradeflow_imf_o","tradeflow_imf_d"]]
df_combined = df_countries.merge(df_unga, left_on='country', right_on='StateName2', how='outer').merge(df_grav, left_on=['country_id','year'], right_on=['country_id_d','year'], how='right')
df_combined=df_combined.drop(columns=["country_id","country"])
df_combined['StateName1']=df_combined['StateName1'].fillna('Singapore')
df_combined['StateName2']=df_combined['StateName2'].fillna(df_combined['country_id_d'].map(df_countries.set_index('country_id')['country']))
df_sipri_sg=df_sipri[(df_sipri["Recipient"]=="Singapore") | (df_sipri["Supplier"]=="Singapore")][["Recipient","Supplier","Year of order"]].rename(columns={"Year of order":"Year"})
df_sipri_sg["Year"]=df_sipri_sg["Year"].astype(int)
df_sipri_sg["country"] = np.where(df_sipri_sg["Recipient"] != "Singapore",df_sipri_sg["Recipient"], df_sipri_sg["Supplier"])
df_sipri_sg=df_sipri_sg.drop(columns=["Recipient","Supplier"])
df_sipri_sg["mil"]=np.ones(len(df_sipri_sg))
df_combined=df_combined.merge(df_sipri_sg, left_on=["StateName2","year"], right_on=["country","Year"], how="left")
df_combined=df_combined.drop(columns=["Year","country"])
df_combined['contig']=df_combined['contig'].fillna(0)
df_combined['comrelig']=df_combined['comrelig'].fillna(0)
df_combined['comlang_off']=df_combined['comlang_off'].fillna(0)
df_combined['comlang_ethno']=df_combined['comlang_ethno'].fillna(0)
df_combined['fta_wto']=df_combined['fta_wto'].fillna(0)
df_combined['diplo_disagreement']=df_combined['diplo_disagreement'].fillna(0)
df_combined['mil']=df_combined['mil'].fillna(0)
df_combined["year"]=df_combined["year"].astype(int)
df_combined=df_combined.dropna(subset=['IdealPointDistance'])
df_combined["contig"]=df_combined["contig"].astype(int)
df_combined["fta_wto"]=df_combined["fta_wto"].astype(int)
df_combined["mil"]=df_combined["mil"].astype(int)
df_combined["comlang_off"]=df_combined["comlang_off"].astype(int)
df_combined["comlang_ethno"]=df_combined["comlang_ethno"].astype(int)

df_vdem_subset=df_vdem[["country_name","histname","country_text_id","year","v2x_polyarchy","v2x_libdem","v2x_partipdem","v2x_delibdem","v2x_egaldem"]]
v2x_columns = ["v2x_polyarchy", "v2x_libdem", "v2x_partipdem", "v2x_delibdem", "v2x_egaldem"]
df_vdem_o = df_vdem_subset.rename(columns={col: col + "_o" for col in v2x_columns})
df_combined = df_combined.merge(df_vdem_o, left_on=["country_id_o", "year"], right_on=["country_text_id", "year"], how="left").drop(columns=["country_name","histname","country_text_id"])
df_vdem_d = df_vdem_subset.rename(columns={col: col + "_d" for col in v2x_columns})
df_combined = df_combined.merge(df_vdem_d, left_on=["country_id_d", "year"], right_on=["country_text_id", "year"], how="left").drop(columns=["country_name","histname","country_text_id"])
df_combined["v2x_polyarchy_d"]=df_combined["v2x_polyarchy_d"].fillna(0)
df_combined["v2x_libdem_d"]=df_combined["v2x_libdem_d"].fillna(0)
df_combined["v2x_partipdem_d"]=df_combined["v2x_partipdem_d"].fillna(0)
df_combined["v2x_delibdem_d"]=df_combined["v2x_delibdem_d"].fillna(0)
df_combined["v2x_egaldem_d"]=df_combined["v2x_egaldem_d"].fillna(0)
df_combined["v2x_polyarchy_o"]=df_combined["v2x_polyarchy_o"].fillna(0)
df_combined["v2x_libdem_o"]=df_combined["v2x_libdem_o"].fillna(0)
df_combined["v2x_partipdem_o"]=df_combined["v2x_partipdem_o"].fillna(0)
df_combined["v2x_delibdem_o"]=df_combined["v2x_delibdem_o"].fillna(0)
df_combined["v2x_egaldem_o"]=df_combined["v2x_egaldem_o"].fillna(0)
df_combined=df_combined.dropna(subset=["gdp_d","pop_d"])
#df_combined["gdp_o"]=df_combined["gdp_o"].fillna(0)
#df_combined["gdp_d"]=df_combined["gdp_d"].fillna(0)
#df_combined["pop_o"]=df_combined["pop_o"].fillna(0)
#df_combined["pop_d"]=df_combined["pop_d"].fillna(0)

df_combined = df_combined.merge(df_tgdp_ratio[["Country Code","Year","Value"]].rename(columns={"Value":"tgdp_ratio_o"}), left_on=["country_id_o","year"], right_on=["Country Code","Year"], how="left").drop(columns=["Country Code","Year"])
df_combined = df_combined.merge(df_tgdp_ratio[["Country Code","Year","Value"]].rename(columns={"Value":"tgdp_ratio_d"}), left_on=["country_id_d","year"], right_on=["Country Code","Year"], how="left").drop(columns=["Country Code","Year"])
df_combined = df_combined.merge(df_tot[["Country Code","Year","adjusted_ratio"]].rename(columns={"adjusted_ratio":"tot_o"}), left_on=["country_id_o","year"], right_on=["Country Code","Year"], how="left").drop(columns=["Country Code","Year"])
df_combined = df_combined.merge(df_tot[["Country Code","Year","adjusted_ratio"]].rename(columns={"adjusted_ratio":"tot_d"}), left_on=["country_id_d","year"], right_on=["Country Code","Year"], how="left").drop(columns=["Country Code","Year"])

df_combined["adjusted_tradeflow"] = df_combined[["tradeflow_comtrade_o","tradeflow_comtrade_d","tradeflow_imf_o","tradeflow_imf_d"]].mean(axis=1, skipna=True)
df_combined=df_combined.drop(columns=["gdp_ppp_o","gdp_ppp_d","gdpcap_ppp_o","gdpcap_ppp_d","tradeflow_baci"])
df_combined=df_combined.dropna(subset=["adjusted_tradeflow"])
df_combined['adjusted_v2x_polyarchy'] = abs(df_combined['v2x_polyarchy_o'] - df_combined['v2x_polyarchy_d'])
df_combined['adjusted_v2x_libdem'] = abs(df_combined['v2x_libdem_o'] - df_combined['v2x_libdem_d'])
df_combined['adjusted_v2x_partipdem'] = abs(df_combined['v2x_partipdem_o'] - df_combined['v2x_partipdem_d'])
df_combined['adjusted_v2x_delibdem'] = abs(df_combined['v2x_delibdem_o'] - df_combined['v2x_delibdem_d'])
df_combined['adjusted_v2x_egaldem'] = abs(df_combined['v2x_egaldem_o'] - df_combined['v2x_egaldem_d'])

#correction for sg-my contig
df_combined["contig"] = df_combined.apply(lambda row: 1 if (row["country_id_o"] == "SGP" and row["country_id_d"] == "MYS.2") else 0, axis=1)

df_combined
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

cols=['StateName1', 'StateName2', 'year', 'IdealPointDistance', 'country_id_o', 'country_id_d', 'contig', 'distcap', 'dist', 
      'diplo_disagreement', 'comlang_off', 'comlang_ethno', 'comrelig', 'pop_o', 'pop_d', 'gdp_o', 'gdp_d', 'gdpcap_o', 'gdpcap_d', 
      'fta_wto', 'tradeflow_comtrade_o', 'tradeflow_comtrade_d', 'tradeflow_imf_o', 'tradeflow_imf_d', 'mil', 'v2x_polyarchy_o', 
      'v2x_libdem_o', 'v2x_partipdem_o', 'v2x_delibdem_o', 'v2x_egaldem_o', 'v2x_polyarchy_d', 'v2x_libdem_d', 'v2x_partipdem_d', 
      'v2x_delibdem_d', 'v2x_egaldem_d', 'tgdp_ratio_o', 'tgdp_ratio_d', 'tot_o', 'tot_d', 'adjusted_tradeflow', 
      'adjusted_v2x_polyarchy', 'adjusted_v2x_libdem', 'adjusted_v2x_partipdem', 'adjusted_v2x_delibdem', 'adjusted_v2x_egaldem']

X_cols=[cols[3]]+cols[6:20]+[cols[24]]+cols[40:]
Y_cols=cols[35:40]

X = df_combined[X_cols]
Y = df_combined[Y_cols]

# Standardize the data
X_scaler = StandardScaler()
X_scaled = X_scaler.fit_transform(X)
Y_scaler = StandardScaler()
Y_scaled = Y_scaler.fit_transform(Y)


df_X_scaled = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)
df_Y_scaled = pd.DataFrame(Y_scaled, columns=Y.columns, index=Y.index)
df_scaled = pd.concat([df_X_scaled,df_Y_scaled],axis=1)

# Compute correlation matrix
corr_matrix = df_scaled.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(15,10))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

X1 = sm.add_constant(df_X_scaled[['IdealPointDistance']])
X2 = sm.add_constant(df_X_scaled)

# Compute VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X2.columns
vif_data["VIF"] = [variance_inflation_factor(X2.values, i) for i in range(X2.shape[1])]

print(vif_data)
```

```python
# Define dependent variable (Trade Flow)
y1 = np.log(df_Y_scaled[["adjusted_tradeflow"]] + 1)  # Log-transform
y2 = df_Y_scaled[["adjusted_tradeflow"]] # non-Log-transform

# Fit regression model
model11 = sm.OLS(y1, X1).fit()
model21 = sm.OLS(y2, X1).fit()
model12 = sm.OLS(y1, X2).fit()
model22 = sm.OLS(y2, X2).fit()

# Display results
print(model11.summary())
print(model21.summary())
print(model12.summary())
print(model22.summary())
```

```python
# Apply PCA
pca = PCA(n_components=len(X2.columns))  # Get all components
pca.fit(X2)

# Explained variance ratios
explained_variance = pca.explained_variance_ratio_

# Determine the number of components needed to explain sufficient variance (e.g., 80%)
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
n_components_needed = np.argmax(cumulative_variance >= 0.80) + 1  # First index where variance >= 80%

# Re-run PCA with selected number of components
pca_refined = PCA(n_components=n_components_needed)
X2_trans = pca_refined.fit_transform(X2)

# Component loadings (weights)
component_weights = np.abs(pca_refined.components_).sum(axis=0)

# Normalize weights to sum to 1
normalized_weights = component_weights / component_weights.sum()

#print(X2)
#print(normalized_weights)

# Output the weights assigned to each variable
weights_df = pd.DataFrame({
    "Variable": X2.columns[:],
    "PCA_Weight": normalized_weights
})

# Construct the new Geopolitical Distance Index (GDI)
X2["GDI"] = np.dot(X2, normalized_weights)

n_components_needed, weights_df
```

```python
df_combined['linear_gdi']=(X2['GDI']+1)/4*100
df_combined['exp_gdi']=2*np.exp(X2['GDI']+1)
df_combined
```
